```python
import PyPDF2
import pyttsx3

pdfoku = open("örnek.pdf","rb") #"örnek" isimli pdf dosyasını açar.

pdfokuyucu = PyPDF2.PdfFileReader(pdfoku)

text = ""

for sayfanumarası in range(pdfokuyucu.numPages):
    sayfa = pdfokuyucu.getPage(sayfanumarası)
    text += sayfa.extractText()
pdfoku.close()

print(text)

çalıştır = pyttsx3.init() #burada "init" okunacak metni temsil eder.
çalıştır.say(text)
çalıştır.runAndWait()
```
